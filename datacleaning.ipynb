```python
# Importing Dependencies
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Bringing in Uncleaned CSV's

csv1 = Path("../Module 18 challenge/Resources/201312-citibike-tripdata_1.csv")
csv2 = Path("../Module 18 challenge/Resources/201412-citibike-tripdata_1.csv")
csv3 = Path("../Module 18 challenge/Resources/201512-citibike-tripdata_1.csv")
bike_trip_2013_df = pd.read_csv(csv1)
bike_trip_2014_df = pd.read_csv(csv2)
bike_trip_2015_df = pd.read_csv(csv3)
```

```python
# Changing The Type of Values for the Gender Column
from numpy import int64


bike_trip_2013_df['gender'] = bike_trip_2013_df['gender'].astype(str)
bike_trip_2014_df['gender'] = bike_trip_2014_df['gender'].astype(str)
bike_trip_2015_df['gender'] = bike_trip_2015_df['gender'].astype(str)

# Replace non-numeric values with NaN
bike_trip_2013_df['birth year'].replace('\\N', np.nan, inplace=True)
bike_trip_2014_df['birth year'].replace('\\N', np.nan, inplace=True)
bike_trip_2015_df['birth year'].replace('\\N', np.nan, inplace=True)

# Convert the column to float first to handle NaNs, then to int
bike_trip_2013_df['birth year'] = bike_trip_2013_df['birth year'].astype(float).astype('Int64')
bike_trip_2014_df['birth year'] = bike_trip_2014_df['birth year'].astype(float).astype('Int64')
bike_trip_2015_df['birth year'] = bike_trip_2015_df['birth year'].astype(float).astype('Int64')
```

```python
#replace int values in gender column:

replacement_dict = {
    '1': 'Male',
    '2': 'Female',
    '0': 'Unknown'
}

bike_trip_2013_df['gender'] = bike_trip_2013_df['gender'].replace(replacement_dict)
bike_trip_2014_df['gender'] = bike_trip_2014_df['gender'].replace(replacement_dict)
bike_trip_2015_df['gender'] = bike_trip_2015_df['gender'].replace(replacement_dict)

bike_trip_2013_df.head(100)
```

```python
current_year=2024

Age = current_year - bike_trip_2013_df['birth year']
bike_trip_2013_df['age'] = Age

Age2 = current_year - bike_trip_2014_df['birth year']
bike_trip_2014_df['age'] = Age2

Age3 = current_year - bike_trip_2015_df['birth year']
bike_trip_2015_df['age'] = Age3

```

```python
# Outputting both files to csv in Outputs folder
Output1 = Path("Outputs/12_2013_citibike_tripdata_cleaned.csv", index=False, header=True)
Output2 = Path("Outputs/12_2014_citibike_tripdata_cleaned.csv", index=False, header=True)
Output3 = Path("Outputs/12_2015_citibike_tripdata_cleaned.csv", index=False, header=True)


bike_trip_2013_df.to_csv(Output1)
bike_trip_2014_df.to_csv(Output2)
bike_trip_2015_df.to_csv(Output3)
```
